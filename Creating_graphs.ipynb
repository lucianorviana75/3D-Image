#Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Defining the function

```python
def f(x,y):
  return np.sqrt(9-x**2+y**2)
```

#Discretizing X and Y.

```python
x=np.linspace(-3, 3 ,50)
y=np.linspace(-3, 3, 50)
```

#Cartesian product artifice

```python
x,y = np.meshgrid(x,y)
```

#Defines Z as Z = f(x, y)

```python
z = f(x, y)
```


#indicates that we will make a 3D graph and names the axes



```python
ax = plt.axes(projection = '3d')
ax.set_xlabel('x')
ax.set_ylabel('Y')
ax.set_zlabel('z')
```

#finally, add the function in this case we are plotting in the wireframe version, that is, you will see the mesh in the surface.

```python
ax = plt.axes(projection = '3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_wireframe(x,y,z)
```


#Discretizing the two angle variables

```python
angle = np.linspace(0,2* np.pi,32)
theta,phi = np.meshgrid(angle, angle)
```


#preaching common values for inner and outer radii

```python
r,R = 0.3,1
```

#Parameterization

```python
x = (R + r * np.cos(phi))*np.cos(theta)
y = (R + r * np.cos(phi))*np.sin(theta)
z = r * np.sin(phi)
```

#Creating a 3D project.

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
```

#Limiting the axes

```python
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)
```

#Displayng the figure

```python

ax.plot_surface(x, y, z, rstride=1, cstride=1 ), plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection= '3d')
ax.set_xlim3d(-1, 1)
ax.set_ylim3d(-1, 1)
ax.set_zlim3d(-1, 1)
ax.plot_surface(x,y,z, rstride = 1, cstride = 1), plt.show()
```
